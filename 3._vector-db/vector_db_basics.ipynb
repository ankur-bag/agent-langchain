```python
import chromadb
client = chromadb.Client()
```

```python
collection = client.create_collection("news_articles")
```

```python
collection.add(
    ids= ["id1", "id2", "id3","id4"],
    documents=[
        "Apple is leading in a smart phone game with iphone sales by 35%",
        "Tesla booked a minor profict of 1 billion $ in Q2",
        "Apples are high in fiber, vitamins and various antioxidants",
        "SpaceX got NASA contract worth 10 billion $"
    ]
)
```

```python
results = collection.query(
    query_texts = ["This is a query related to Elon Musk"],
    n_results = 1
)
print(results)
```

```python
results = collection.query(
    query_texts = ["i lvoe ipad"],
    n_results = 2
)
print(results)
```
