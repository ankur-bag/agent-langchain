```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
with open("blood_work.txt", "r") as f:
    blood_report = f.read()
    
print(blood_report[:200])
```

```python
extraction_prompt = f"""
you are a medical data extraction assistant. from the blood report below , extract all test values and classify each one as HIGH , LOW or NORMAL , based on the reference ranges provided in the report. 

Format you response as:
- Test Name: value | Status : HIGH/LOW/NORMAL | Reference : range

Blood Report:
{blood_report}
"""

extraction_response = llm.invoke(extraction_prompt)
extraction_values = extraction_response.text

print("=== STAGE 1 : EXTRACTED VALUES ===")
print(extraction_values)
```

```python
diet_prompt = f"""
    You are a clinical nutritionist specializing in Indian dietary habits.
    
    Based on the blood work analysis below , write:
    1.A short health summary in 3-4 lines explaning the patient's condition in simple language
    2. A short, practical Indian diet plan having only two sections (1) Foods to avoid (2) Foods to eat more of.
    Do not include any other sections in the diet plan
    
    Blood work analysis:{extraction_values}
"""

diet_response = llm.invoke(diet_prompt)
print(diet_response.text)
```
